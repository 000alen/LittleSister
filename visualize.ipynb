```python
import pandas
import geopandas
import folium

from folium import plugins
from shapely.geometry import Point
```

```python
files_paths = [ # A13101: SANTIAGO
    "database\probability_point\A13101_GONZALO_WINTER_ETCHEBERRY_0.03.csv",
    #"database\probability_point\A13111_GONZALO_WINTER_ETCHEBERRY_0.03.csv",
    #"database\probability_point\A13118_GONZALO_WINTER_ETCHEBERRY_0.03.csv",
    #"database\probability_point\A13120_GONZALO_WINTER_ETCHEBERRY_0.03.csv",
    #"database\probability_point\A13123_GONZALO_WINTER_ETCHEBERRY_0.03.csv",
    #"database\probability_point\A13129_GONZALO_WINTER_ETCHEBERRY_0.03.csv"
]
```

```python
df = None
for file_path in files_paths:
    if df is None:
        df = pandas.read_csv(file_path)
    else:
        df = df.append(pandas.read_csv(file_path))

df = df[df["probability"] != 0]

df
```

```python
geometry = geopandas.points_from_xy(df.longitude, df.latitude)
geo_df = geopandas.GeoDataFrame(df["probability"], geometry=geometry)
geo_df
```

```python
view = folium.Map(location=[-33.4,-70.6], zoom_start=11) #location = [15,30], tiles='Cartodb dark_matter', zoom_start = 2)
heat_data = [[point.xy[1][0], point.xy[0][0]] for point in geo_df.geometry]
plugins.HeatMap(heat_data, radius=10).add_to(view)
plugins.FastMarkerCluster(heat_data).add_to(view)
```
